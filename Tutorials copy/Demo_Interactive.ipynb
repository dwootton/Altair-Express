## Interaction Technique Demo

Interaction techniques allow for you to engage data in a conversation. 


This notebooks is used to demo this functionality. 

```python
# imports 
import altair_express as alx 
from vega_datasets import data 
import numpy as np
df = data.cars()
df['year'] = df['Year'].dt.year
```

```python

df = data.gapminder()
df['log_pop'] = np.log(df['pop'])

columns = df.columns

alx.profile(df,vars=columns)

```

```python
alx.hist(df, x= 'log_pop')
```

```python
alx.scatterplot(df,x='life_expect',y='fertility')
```

```python
alx.jointplot(df,x='life_expect',y='fertility')
```

```python
alx.lineplot(df,x='year',y='average(life_expect)', color= 'cluster')
```

```python
alx.stripplot(df[df['year']==2005],x='life_expect',row='cluster',color='cluster:N')
```

```python
correlation_matrix = df.corr()
alx.heatmap(correlation_matrix) # heatmap takes a NxN matrix of numbers
```

```python
alx.pairplot(df)
```

```python
alx.barplot(data.barley(),x='year:N',y='sum(yield)',color='year',column='site')
```

```python
df = data.cars()
```

```python
alx.tooltip_hover() + alx.scatterplot(df,x='Horsepower:N',y='Miles_per_Gallon')
```

```python
alx.pan_zoom() + alx.scatterplot(df,x='Horsepower',y='Miles_per_Gallon')
```

```python
df = data.cars()

alx.tooltip_hover()+ df.query(" (Horsepower>=132.58 and Horsepower<=228.58)  and  (Miles_per_Gallon>=7.37 and Miles_per_Gallon<=22.45) ")
```

Filter Type

```python
          
df = data.gapminder()

alx.filter_type(target='country') + alx.scatterplot(df,x='fertility',y='life_expect')          
```

```python
alx.tooltip_hover()+alx.filter_type(target='country') + alx.scatterplot(df,x='fertility',y='life_expect')          
```

Label point

```python
alx.label_point(field='Name')+alx.scatterplot(data.cars(),x='Miles_per_Gallon',y='Horsepower')
```

Overview + Detail

```python
brush = alx.highlight_brush()
df = data.cars()

overview = brush + alx.hist(df,x='Miles_per_Gallon')

detail = alx.hist(df,x='Miles_per_Gallon', effects={'filter':[brush]})

overview & detail
```

```python
brush = alx.highlight_brush()
df = data.cars()
overview =  alx.hist(df,x='Miles_per_Gallon')
detail = alx.hist(df,x='Miles_per_Gallon')
brush + (overview & detail)
```

Copy Selection to Query

```python
alx.highlight_brush() + alx.jointplot(data.cars(),x='Horsepower',y='Miles_per_Gallon')
```
