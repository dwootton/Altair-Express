# Interaction Effects as arguments

As interaction effects are changes to a spec, we can pass interaction variables into functions to create their specific effects. 

For example, filter brushing removes points from datasets:
## Filter Brush

```python
interaction = alx.filter_brush()

hist1 = interaction + alx.hist(data=data.cars(),x='Horsepower') 
hist1 & alx.hist(data=data.cars(),x='Horsepower', effects={"filter":interaction})
```

## Highlight Brush
Whereas a highlight brush keeps the reference dataset there:

```python
interaction = alx.highlight_brush()
hist2 = interaction + alx.hist(data=data.cars(),x='Horsepower')
#hist2 & 
hist3= interaction + alx.hist(data=data.cars(),x='Miles_per_Gallon')
hist2&hist3
```

and you can even merge them 

```python
interaction = alx.highlight_brush()

hist1 = interaction + alx.hist(data=data.cars(),x='Horsepower') 
hist1 & alx.hist(data=data.cars(),x='Horsepower', effects={'filter':interaction})
```

```python
## interaction = alx.highlight_brush

interaction + alx.scatterplot(data=data.cars(),x='Horsepower',y='Miles_per_Gallon') & alx.hist(data=data.cars(),x='Miles_per_Gallon', highlights=[interaction])
```
