```python
df = df.groupby(['Age','Year']).aggregate('sum').reset_index()

# calculate cumsum of Pop
df['no_csum'] = df.groupby(['Year'])['People'].cumsum()
```

```python
df
column = 'no_csum'

df['norm'] = df.groupby('Year')['no_csum'].transform(lambda x: (x) / x.max())
```

```python
df['people_per_hundred']=df.groupby('Year')['People'].apply(lambda x:round(100 * x / float(x.sum())))
```

```python
 

exploded = df.loc[df.index.repeat(df.people_per_hundred)]
exploded
```

```python
alt.Chart(exploded[exploded['Year']==2000]).mark_circle(opacity=1,size=125,color='lightgray').transform_window(
    id='rank()',
    groupby=['Age']
).encode(
    alt.X('Age:O'),
    alt.Y('id:O',
          axis=None,
          sort='descending')
).properties(height=100).configure_view(
    strokeWidth=0
)
```
